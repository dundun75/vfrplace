```python
import csv

# Inizializza una lista per memorizzare i codici degli articoli
codici_articoli = []

# Apre il file CSV in modalità lettura
with open('tuo_file.csv', newline='') as file:
    reader = csv.reader(file, delimiter=';')
    
    # Legge la prima riga per ottenere i nomi degli articoli
    prima_riga = next(reader)
    nomi_articoli = prima_riga[14:21]  # Dalla colonna 15 alla 20
    codici_articoli = nomi_articoli
    
    # Inizializza una lista per memorizzare i dati di ogni riga
    dati_per_riga = []
    
    # Legge le righe successive
    for row in reader:
        quantita = row[14:21]  # Le quantità corrispondenti iniziano dal campo 15 al 20
        dati_per_riga.append(list(zip(codici_articoli, quantita)))

# Stampa i dati per ogni riga
for idx, dati_riga in enumerate(dati_per_riga):
    print(f"Riga {idx + 1}: {dati_riga}")
```
